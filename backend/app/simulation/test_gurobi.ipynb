```python
import gurobipy as gp
import os 
```

```python
# Crear un modelo
m = gp.Model("ejemplo_optimizacion")

# Agregar variables de decisión
x = m.addVar(vtype=gp.GRB.CONTINUOUS, name="x")
y = m.addVar(vtype=gp.GRB.CONTINUOUS, name="y")

# Establecer la función objetivo
m.setObjective(2 * x + 3 * y, sense=gp.GRB.MAXIMIZE)

# Agregar restricciones
m.addConstr(x + y <= 9, name="restriccion_1")
m.addConstr(x - y >= 1, name="restriccion_2")
m.addConstr(x >= 0, name="restriccion_3")
m.addConstr(y >= 0, name="restriccion_4")

# Optimizar el modelo
m.optimize()

# Imprimir resultados
for v in m.getVars():
    print(f"{v.varName}: {v.x}")

print("Función objetivo:", m.objVal)
```
